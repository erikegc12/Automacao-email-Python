# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
!pip install pyautogui
```

```python
import pyautogui
import pyperclip
import time

pyautogui.PAUSE = 1

# pyautogui.click -> clicar
# pyautogui.press -> apertar 1 tecla
# pyautogui.hotkey -> conjunto de teclas
# pyautogui.write -> escreve um texto

#1 passo vamos acessar o servidor


pyautogui.hotkey("ctrl", "t")
pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing")
pyautogui.hotkey("ctrl","v")
pyautogui.press("enter")
time.sleep(3)


#2, 3 passo, vamos logar e fazer o donwload da base de dados 
pyautogui.click(x=405, y=300, clicks=2)
time.sleep(4)
pyautogui.click(x=415, y=369)
pyautogui.click(x=1721, y=193)
pyautogui.click(x=1500, y=620)

#4 IMPORTANDO VALORES COM PANDAS AS PD !!!!IPC!!!
#lembre-se de colocar o .xlsx ou qualquer coisa que o arquivo seja, html ou sei lá e definir a tabela como VAR
import pandas as pd
tabela = pd.read_excel(r"C:\Users\Erike\Intensivo Python\Aula 01\Vendas-dezembro.xlsx")
display(tabela)

#5 calcular os indicadores COMANDOS PARA CALCULAR AS TABELAS DO ARQUIVO
faturamento = tabela["Valor Final"].sum()
print(faturamento)
quantidade = tabela["Quantidade"].sum()
print(quantidade)

#6 vamos enviar um email
#abrindo o email e esperando o tempo
pyautogui.hotkey("Ctrl","t")
pyperclip.copy("https://mail.google.com/mail/u/0/#inbox")
pyautogui.click(x=374, y=46)
pyautogui.hotkey("Ctrl","v")
pyautogui.press("enter")
time.sleep(4)

#Enviando o E-mail com suas informações
pyautogui.click(x=209, y=189)
pyperclip.copy("pythonimpressionador@gmail.com")
pyautogui.hotkey("ctrl","v")
pyautogui.press("tab")

pyperclip.copy("Área de Vendas")
pyautogui.hotkey("ctrl","v")
pyautogui.press("tab")

# formatamos o texto aqui para sere enviado exatamente como está
texto = f"""
Prezado cliente, Boa tarde! Venho por meio diponibilizar o relatório de vendas e quantidade.

Aqui seguem os dados:
Qtd vendas: {quantidade:,.2f}
Faturamento em reais: R$:{faturamento:,}

Qualquer dúvida estou à disposição.
Att.,
Erik Correa
"""

#aqui pegamos a VAR texto e inserimos aqui, fazendo ela ser copiada e enviando efetivamente a mensagem.
pyperclip.copy(texto)
pyautogui.hotkey("ctrl","v")
pyautogui.hotkey("ctrl","enter")

#agora vamos verificar a mensagem enviada
pyautogui.click(x=198, y=355)
pyautogui.click(x=536, y=285)
```
